```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
def concat_ones(data):
    
    size = len(data)
    new_data = np.concatenate((data, np.ones((size, 1))), axis=1)
    
    return new_data

def prepare_data():
    
    uv = concat_ones(np.load("15_uv.npy"))
    xyz = concat_ones(np.load("15_XYZ.npy") * 7)

    return uv, xyz
```

```python
# Import all inputs
uv, xyz = prepare_data()
img = plt.imread("stereo2012a.jpg")
```

```python
# Show image

fig, ax = plt.subplots(1, 2, figsize=(15,7))

ax[0].imshow(img)
ax[0].set_title("Image (stereo2012a) for Camera Calibration")

ax[1].imshow(img)
ax[1].scatter(uv[:,0], uv[:,1], s=2, c='w')
ax[1].set_title("Selected points for Calibration")
plt.show()
```

```python
# Determine matrix equations A
a = []
for i, j in zip(uv, xyz):
    
    u, v = i[:2]
    x, y, z = j[:3]
    
    equations = np.array([[0, 0, 0, 0, -x, -y, -z, -1, v*x, v*y, v*z, v],
                          [x, y, z, 1, 0, 0, 0, 0, -u*x, -u*y, -u*z, -u]])
    
    a.append(equations)

A = np.array(a).reshape((-1,12))
```
