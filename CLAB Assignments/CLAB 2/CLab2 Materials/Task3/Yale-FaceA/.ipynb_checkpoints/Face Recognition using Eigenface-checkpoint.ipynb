```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
# List of filepaths for all training images
im_filepaths = glob.glob("trainingset/*")

image_shape = 231, 195

# Add all image matrices to a numpy array to create a big data matrix of images
training_images = np.array([plt.imread(image).flatten() for image in im_filepaths])

training_images.shape
```

```python
# Determine the average of all the training face images
avg_face = training_images.mean(axis=0)

# Plot the average of all face images
plt.imshow(avg_face.reshape(image_shape))
plt.title("Average of Training Images")
plt.show()
```

```python
# Normalise the training data by subtracting the mean image from each one
normalised_training_images = training_images - avg_face
normalised_training_images.shape


# Display a sample of the normalised images for each subject
fig, ax = plt.subplots(2, 7, figsize=(30,10))

for i in np.arange(0, 140, 10):
    
    sample_im = normalised_training_images[i,:].reshape(image_shape)
    
    ax[0 if i < 70 else 1, i%7].imshow(sample_im, cmap='gray')
 
fig.suptitle("Samples of Normalized Images", fontsize = 16)
plt.show()
```

```python
# Calculate the covariance matrix of the normalised images
cov_matrix = np.cov(normalised_training_images)

# Calculate the eigen-values and eigen-vectors using the covariance matrix
eig_values, eig_vectors = np.linalg.eig(cov_matrix)
```

```python
# Sort eigenvalues and eigenvectors (wrt descending order of eigenvalues) for ordered Principal Components
desc_sorted_index = eig_values.argsort()[::-1]

eig_values = eig_values[desc_sorted_index]  # Sorted eigenvalues
eig_vectors = eig_vectors[:, desc_sorted_index] # Corresponding eigenvectors

# Determine the percentage of Variance Explained by the Principal Components
explained_variance_perc = eig_values / np.sum(eig_values) * 100
cumulative_explained_variance_perc = np.cumsum(explained_variance_perc)

# Visualize the explained variance percentages
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,8))

# Plot the Explained Variance Percentage for the top 25 Principal Components
ax1.bar(np.arange(1,26), explained_variance_ratio[:25])
ax1.set_title("Variance Explained by Top 25 Principal Components", fontsize=16)
ax1.set_ylabel("Explained Variance Percentage", fontsize=14)
ax1.set_xlabel("# Principal Component", fontsize=14)

# Plot the Cumulative Explained Variance Percentage for all Principal Components
ax2.scatter(np.arange(len(eig_values)), cumulative_explained_variance_ratio)
ax2.hlines(y=80, xmin=1, xmax=50)   # To determine how many PCs explain 80% of the variance in the data
ax2.vlines(x=15, ymin=1, ymax=100)  # Found to be 18
ax2.set_title("Cumulative Explained Variance", fontsize=16)
ax2.set_ylabel("Cumulative Explained Variance Percentage", fontsize=14)
ax2.set_xlabel("# Principal Component", fontsize=14)

fig.suptitle("Determining k such that k Principal Components explain 80% of the Variance in the Data", fontsize=20)
fig.savefig("Determining K.png")

plt.show()

# Print the explained variance percentage for first k = 15 Principal Components
perc_k = np.real(np.round(cumulative_explained_variance_perc[14] ,2))
print("Explained variance percentage for first 15 Principal Components = {}%".format(perc_k))
```

```python
# Assign k, where k is the number of Principal Components
k = 15

# Select the first k Principal Components (eigen-vectors)
k_components = eig_vectors[:k, :]

# Determine the eigen-faces using the first k Principal Components
eigen_faces = k_components @ normalised_training_images

# Visualise the top k eigen-faces
fig, ax = plt.subplots(3, 5, figsize=(20,10))

for i, face in enumerate(eigen_faces):    
    
    ax_idx = i // 5, i % 5
    
    ax[ax_idx].imshow(np.real(face.reshape(image_shape)), cmap='gray')
    ax[ax_idx].set_title("Eigenface %.0f"%(i+1))
    ax[ax_idx].set_xticks([])
    ax[ax_idx].set_yticks([])

fig.suptitle("Top k = {} Eigenfaces".format(k), fontsize=16)
plt.show()
```

```python
# Project eigen-faces onto k-dimensional subspace of eigen-vectors
model_weights = normalised_training_images @ eigen_faces.T
```

```python
test_data_filepaths = glob.glob("testset/*")

# Vectorise test images
test_images = np.array([plt.imread(image).flatten() for image in test_data_filepaths])

# Normalise the test data by subtracting the average image from each one
normalised_test_images = test_images - avg_face

# Project test images onto the k Principal Component subspace
test_weights = normalised_test_images @ eigen_faces.T

# Calculate distance of all test images from each training image in the new subspace
dist_from_train_images = np.linalg.norm(test_weights[:,None] - model_weights, axis=2) 

# Determine the indexes of 3 "most similar" training images (those having least distances from each test image)

# a) Sorted indices of training images in order of decreasing similarity to test image
test_preds = dist_from_train_images.argsort(axis=1)  
# b) Select 3 most similar images (corresponding to first 3 indices)
test_preds = [test_preds[i][:3] for i in range(len(test_preds))]
```

```python
# Visualize predictions
fig, ax = plt.subplots(10, 4, figsize=(15,42))
for i, test_img in enumerate(test_images):
#     fig, ax = plt.subplots(1, 4, figsize=(15,7.5))
    
    most_similar_train_image = training_images[test_preds[i]]

    ax[i,0].imshow(test_img.reshape(image_shape), cmap='gray')
    ax[i,0].set_title("Test Sample {}".format(i+1))
    ax[i,0].set_xticks([])
    ax[i,0].set_yticks([])
    ax[i,0].spines['bottom'].set_color('1')
    ax[i,0].spines['top'].set_color('1')
    ax[i,0].spines['left'].set_color('1')
    ax[i,0].spines['right'].set_color('1')

    ax[i,1].imshow(most_similar_train_image[0].reshape(image_shape), cmap='gray')
    ax[i,1].set_title("#1 Prediction")
    ax[i,1].set_xticks([])
    ax[i,1].set_yticks([])
    
    ax[i,2].imshow(most_similar_train_image[1].reshape(image_shape), cmap='gray')
    ax[i,2].set_title("#2 Prediction")
    ax[i,2].set_xticks([])
    ax[i,2].set_yticks([])
    
    ax[i,3].imshow(most_similar_train_image[2].reshape(image_shape), cmap='gray')
    ax[i,3].set_title("#3 Prediction")
    ax[i,3].set_xticks([])
    ax[i,3].set_yticks([])

fig.savefig("a.png")
plt.show()
```
