#### Some cleaning functions

```python
def transform_tstrength(data):
    
    # Code to clean the Tensile Strength Column
    new_data = data.copy()
    
    new_data = new_data.dropna(axis=0)
    
    noise_elements = [' \xa0', ' Min', ' Typ', 'Typ', ' Max', ' for Standard', 'Tensile Strength', '', " ", "   "]
    
    # Remove noise elements from observations
    for ele in noise_elements:
        new_data['Tensile Strength'] = new_data['Tensile Strength'].apply(lambda x: x.replace(ele, ""))
    # Remove trailing whitespace
    new_data['Tensile Strength'] = new_data['Tensile Strength'].apply(lambda x: x.strip())
    # Convert to numeric type
    new_data['Tensile Strength'] = new_data['Tensile Strength'].astype('float')
    
    return new_data

def temper_mapping(data, return_mapping=False):
    
    new_data = data.copy()    
    new_data['Temper'] = new_data['Temper'].fillna('NA')
    
    temper_map = {}
    x = set(list(zip(new_data['Temper'], new_data['Temper Code'])))
    
    for name, code in x:
        if name != 'NA':
            temper_map[code] = name 

    new_data['Temper'] = new_data['Temper Code'].apply(lambda x: temper_map[x])
    
    
    if return_mapping:
        return new_data, temper_map
    else:
        return new_data
    
def transform_form(data):
    
    # Code to clean the Form column
    new_data = data.copy()
    
    form_cleaning_list = ["flat products", 'flatproducts', 'flatt products'] # list of items to replace with "products"
    
    form_clean = new_data['Form'].apply(lambda x: x.lower()) # convert to lowercase
    form_clean = form_clean.str.strip() # remove extra whitespaces    

    for form_type in form_cleaning_list:
        form_clean = form_clean.replace( form_type , "products")
        
    new_data['Form'] = form_clean
        
    return new_data
```

```python
data = pd.read_csv("Copper alloy data set - Book2 (version 3).csv")
hidden_data = data.sample(40)

data = data[~data.index.isin(hidden_data.index)]

data.info()
```

**1. Clean Tensile Strength Column**

The type of the column was object indicating the presence of some strings or missing values. We will use the function transform_tstrength to remove unnecessary prefixes, whitespace and convert to numerical type.

```python
df = transform_tstrength(data)
df['Tensile Strength'].describe()
```

**2. Clean Form Column**

The form column consists of redundant noisy elements such as "flatt products", "flatproducts", " flat products". We can simply classify these as "products". Using our function transform_form, we will clean this column.

```python
df = transform_form(df)
df['Form'].unique()
```

**3. Clean Temper and Temper Code columns**

The Temper had some missing values. These records with missing "temper" fortunately had non-null values for 'temper code'. Therefore the missing values in the temper column were filled using the 'temper code' --> 'temper' mapping.

The mapping looked something like this:

- **OS050** --> Nominal Grain Size 0.050 mm
- **TH04** --> Hard and Precipitation Heat Treated
- **H00** --> 1/8 Hard

and so on ...



```python
df = temper_mapping(df)
df.info()
```

```python
complete_df = df.dropna(axis=0)
complete_df.info()
```

**3. Remove duplicate rows**

The dataset contains 82 duplicate records (refer 1. EDA.ipynb). We will go on and remove these redundant records. By doing this we attempt to diminish the possibility of introducing bias in the model, which might arise due to duplicate records.

```python
complete_df_without_duplicates = complete_df[~complete_df.duplicated(keep='first')]
complete_df_without_duplicates.info()
```

```python
X = complete_df_without_duplicates.drop(['Tensile Strength', 'Thermal ConductivityBtu/ sq ft/ ft hr/ °F at 68°F'], axis=1)
y = complete_df_without_duplicates[['Tensile Strength', 'Thermal ConductivityBtu/ sq ft/ ft hr/ °F at 68°F']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X_train, X_test = X_train.select_dtypes('number'), X_test.select_dtypes('number')
model = RandomForestRegressor()
model = model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

mse = mean_squared_error(y_train, train_pred)
mae = mean_absolute_error(y_train, train_pred)
r2 = r2_score(y_train, train_pred)

print("Train :",mse, mae, r2)

mse = mean_squared_error(y_test, test_pred)
mae = mean_absolute_error(y_test, test_pred)
r2 = r2_score(y_test, test_pred)

print("Test :",mse, mae, r2)

print()
importances = model.feature_importances_
std = np.std([tree.feature_importances_ for tree in model.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")
labels = []
for f in range(X_train.shape[1]):
#     print("%d. feature %s (%f)" % (f + 1, X_train.columns[indices[f]], importances[indices[f]]))
    labels.append(X_train.columns[indices[f]])

# Plot the feature importances of the forest
plt.figure(figsize=(20,10))
plt.title("Feature importances")
plt.bar(labels, importances[indices],
       color="r", yerr=std[indices], align="center")
# plt.xticks(ticks=labels)
plt.xlim([-1, X.shape[1]])
plt.show()
```
