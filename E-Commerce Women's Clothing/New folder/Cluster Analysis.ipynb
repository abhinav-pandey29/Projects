```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

## Products Worth Inspecting

1. From K-Means - Cluster 3 (For Disliked Products)
2. From Agglomerative Clustering - Clusters 4 (For Disliked Products)

```python
df = pd.read_csv("Clustering Results.csv")
```

```python
k3 = df[df['cluster'] == 3]
a4 = df[df['cluster_agg'] == 4]
```

**1. Class Analysis**

```python
def analyse_cluster_no_normalize(col_name, sharey=True):

    columns = df.columns[df.columns.str.contains(col_name)]

    fig, ax = plt.subplots(1,3, figsize=(20,5), sharey=sharey)

    class_k3 = k3[columns]
    class_k3.sum().plot(kind='bar', ax=ax[0])
    ax[0].set_title("K-Means Cluster 3")

    class_a4 = a4[columns]
    class_a4.sum().plot(kind='bar', ax=ax[1])
    ax[1].set_title("Agglomerative Cluster 4")

    df_all = df[columns]
    df_all.sum().plot(kind='bar', ax=ax[2])
    ax[2].set_title("Overall Data")

    plt.show()
```

```python
def analyse_cluster(col_name, normalise=True, ylim=True):

    columns = df.columns[df.columns.str.contains(col_name)]

    fig, ax = plt.subplots(1,3, figsize=(20,5))
    
    if normalise:
        class_k3 = k3[columns].sum() / df[columns].sum() * 100
        class_a4 = a4[columns].sum() / df[columns].sum() * 100
    else:
        class_k3 = k3[columns].sum()
        class_a4 = a4[columns].sum()
        
    class_k3.plot(kind='bar', ax=ax[0])
    ax[0].set_title("K-Means Cluster 3")
    
    class_a4.plot(kind='bar', ax=ax[1])
    ax[1].set_title("Agglomerative Cluster 4")
    if ylim:
        ax[0].set_ylim([0,105])
        ax[1].set_ylim([0,105])

    df_all = df[columns]
    df_all.sum().plot(kind='bar', ax=ax[2])
    ax[2].set_title("Overall Data")

    plt.show()
```

```python
analyse_cluster("Class Name")
```

**2. Department Analysis**

```python
analyse_cluster("Department Name", ylim=False)
```
