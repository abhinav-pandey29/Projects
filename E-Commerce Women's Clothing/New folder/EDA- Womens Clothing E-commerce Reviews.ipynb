```python
subset = df[['Clothing ID', 'Rating']]

rating_cols = pd.get_dummies(subset['Rating'])

ratings_df = pd.concat((subset, rating_cols), axis=1).drop('Rating', axis=1)
ratings_df = ratings_df.groupby('Clothing ID').sum().astype(int)

ratings_df['total_ratings'] = ratings_df.sum(axis=1)

ratings_df
```

```python
# Visualize the difference in average ratings across all the different clothing classes

averages_by_classname = data.groupby('Class Name').mean()

average_rating_by_classname = averages_by_classname['Rating']

average_rating_by_classname.plot(kind='bar')
```

```python
# Visualize the difference in average ratings across all the different clothing classes

##### DATA PREPARATION FOR VISUALIZATION #####
averages_by_classname = data.groupby('Class Name').mean()

average_rating_by_classname = averages_by_classname['Rating']

sorted_avg_rating_by_classname = average_rating_by_classname.sort_values()
################ END ##############

######### VISUALIZATION CODE ############
plt.figure(figsize=(10, 6))
sorted_avg_rating_by_classname.plot(kind='barh')
plt.xlabel("Ratings")
plt.title("Average Rating for Each of the Product Classes")
plt.xticks(range(0,6))
plt.show()
```

```python
# Visualize the difference in average ratings across all the different customer ages

##### DATA PREPARATION FOR VISUALIZATION #####

averages_by_age = data.groupby('Age').mean()

average_rating_by_age = averages_by_age['Rating']

################ END ##############

######### VISUALIZATION CODE ############

plt.figure(figsize=(15, 6))
average_rating_by_age.plot()
plt.xlabel("Age")
plt.title("Average Rating by all Customer Ages")
plt.yticks(range(0,6))
plt.show()
```
