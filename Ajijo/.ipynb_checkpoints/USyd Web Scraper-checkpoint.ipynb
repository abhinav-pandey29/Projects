```python
def test_run(course_codes):
   
    print("Total Courses = {}".format(len(course_codes)))
    success=0
    
    failed_urls = []
    
    course_details = pd.DataFrame()
    
    for code_idx, code in enumerate(course_codes):
        
        clear_output(wait=True)
        try:

            url = "https://www.sydney.edu.au/units/{}/2020-S2C-ND-CC".format(code)    
            page = requests.get(url)
            soup = BeautifulSoup(page.content, 'html.parser')

            if "Error (404)" in soup.find_all('title')[0].text:

                url = url.replace("S2C", "S1C")
                page = requests.get(url)
                soup = BeautifulSoup(page.content, 'html.parser')

            course_code, course_name = soup.find_all('h1', class_="pageTitle b-student-site__section-title")[0].text.split(":")
            course_name = course_name.strip()

            semester_info = soup.find_all('div', class_='breadcrumbsModule')[0].text.strip().split("\n")[2]

            assessment_table = soup.find_all('table', id="assessment-table")
            learning_outcomes = soup.find_all('div', id="learningOutcomes")
            learning_outcomes = learning_outcomes[0].text

            types_ = [x.find_all('img')[0].get('alt').split()[0].title() + " " + x.text.strip() 
                  if (len(x.find_all('img')) > 0) 
                  else x.text.strip() 
                  for x in assessment_table[0].find_all("td", rowspan="2")]
            
            n_assessments = len(types_)
            
            hurdle_or_not = []
            for i in assessment_table[0].find_all("td", rowspan="2"):
                if len(i.find_all('img')) > 1:
                    hurdle_or_not.append(i.find_all('img')[1].get('alt').split()[0].title())
                else:
                    hurdle_or_not.append("'n/a'")
            
            assessment_details = assessment_table[0].find_all('td', rowspan="")
            weightage = [assessment_details[i].text for i in range(1, 5 * n_assessments + 1, 5)]
            length = [assessment_details[i].text for i in range(3, 5 * n_assessments + 3, 5)]
            outcomes_assessed = [assessment_details[i].text for i in range(4, 5 * n_assessments + 3, 5)]
            outcomes_assessed = [outcome.replace("Outcomes assessed:", "").replace("\n", "").strip() for outcome in outcomes_assessed]
            outcomes_assessed = [outcome.replace("L", " L").strip() for outcome in outcomes_assessed]
            descriptions = [assessment_details[i].text.strip() for i in range(0, 5 * n_assessments, 5)]
            due = [assessment_details[i].text.strip() for i in range(2, 5 * n_assessments + 1, 5)]  
            
            s = soup.find_all('div', id='weekly_schedule_panel')[0].find_all('td', class_="", rowspan="")
            p = r"\D*\d+\.*\d* hr\)"    
            activity = [re.findall(p, ele.text)[0] for ele in s if len(re.findall(p, ele.text)) > 0]
            activity = np.unique(activity)
            
            success += 1
            
            print("Progress : {} %".format(np.round((code_idx+1)/len(course_codes)*100, 2)))
            
            cd = pd.DataFrame({"Course Code" : [course_code] * n_assessments,
                                  "Course Name" : [course_name] * n_assessments,
                                  "Semester" : [semester_info] * n_assessments,
                                  "Activity" : [activity] * n_assessments,
                                  "Assignments" : types_,
                                  "Hurdle or Not" : hurdle_or_not,
                                  "Weightage" : weightage,
                                  "Duration" : length,
                                  "Outcomes" : outcomes_assessed,
                                  "Descriptions" : descriptions,
                                  "Due" : due,
                                  "Course Objectives" : [learning_outcomes] * n_assessments})
            
            course_details = pd.concat((course_details, cd), axis=0)
            
            
        except:
            failed_urls.append(url)
            continue
        
    print("Number of Successful Cases Run = {}".format(success))
    
    return course_details.reset_index(drop=True), failed_urls
```

```python
course_codes = {'MTRX' : [1701, 1702, 1705, 2700, 3700, 3760, 5700],
                'AMME' : [1802, 2500, 1362, 2301, 3500, 2000, 2200, 4710, 5790, 5520, 3060],
                'MECH' : [2400, 3460, 5720, 3660],
                'ELEC' : [1103, 3204, 2104, 3404],
                'MATH' : [2021, 2921]}

all_courses = [k + str(code) for k, codes in course_codes.items() for code in codes]

course_details, failed_urls = test_run(all_courses)
```
