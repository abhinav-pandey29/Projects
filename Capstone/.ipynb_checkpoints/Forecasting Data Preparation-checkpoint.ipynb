```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from glob import glob
```

```python
# List of filenames containing our FOREX data
forex_tables_glob = glob("data/*.xls")
forex_tables_glob
```

```python
def create_forex_df(all_files_glob):
    from glob import glob

    all_files = all_files_glob

    li = []

    for filename in all_files:
        tdf = pd.read_excel(filename, skiprows=5, na_values=['CLOSED', 'closed', 'Closed']).drop([0,1,2,3,4])
        tdf = tdf.rename(columns={'Units':'Date'})

        # Convert Date to Datetime type
        tdf['Date'] = pd.to_datetime(tdf['Date'])

        # Create time series index
        tdf = tdf.set_index('Date', drop=True)
        li.append(tdf)

    df = pd.concat(li, axis=0, sort=False)
    
    return df

data = create_forex_df(forex_tables_glob)
```

```python
def plot_currency_against_AUD(A, data=data):
    B = 'AUD'
    fig, (ax1, ax2) = plt.subplots(2,1, sharex=True)

    data[A].astype('float').plot(ax=ax1)
    ax1.set_title("{}/{} trend since 1984".format(A,B))
    ax1.set_ylabel("Value of {} (per {})".format(A,B))

    aud_per_usd = (1 / data[A]).astype('float')
    aud_per_usd.plot(ax=ax2)
    ax2.set_title("{}/{} trend since 1984".format(B,A))
    ax2.set_ylabel("Value of {} (per {})".format(B,A))

    plt.subplots_adjust(top=0.92, bottom=0.08, left=-1.0, right=0.95, hspace=0.25,
                        wspace=0.35)
    plt.show()

plot_currency_against_AUD('USD')
plot_currency_against_AUD('JPY')
plot_currency_against_AUD('GBP')
```

```python
from statsmodels.tsa.arima_model import ARMA

y = df['USD']['2012':'2016'].astype('float').dropna()
yt = df['USD']['2017':].astype('float').dropna()

model = ARMA(y, order=(1,10))
results = model.fit()
```

```python
x = pd.DataFrame(zip(yt.index, model.predict(yt)))
x = x.set_index(0)

fig, ax = plt.subplots()

yt.plot(ax=ax)
y.plot(ax=ax)
x.plot(ax=ax, c='black')
```
