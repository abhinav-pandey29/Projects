```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from glob import glob
```

```python
def create_forex_df(all_files_glob=glob("data/*.xls")):

    all_files = all_files_glob

    li = []

    for filename in all_files:
        tdf = pd.read_excel(filename, 
                            skiprows=5,
                            na_values=['CLOSED', 'closed', 'Closed', ' --'])
        tdf = tdf.drop([0,1,2,3,4])
        tdf = tdf.rename(columns={'Units':'Date'})

        # Convert Date to Datetime type
        tdf['Date'] = pd.to_datetime(tdf['Date'])

        # Create time series index
        tdf = tdf.set_index('Date', drop=True)
        li.append(tdf)

    df = pd.concat(li, axis=0, sort=False)
    
    return df
```

```python
# Initialize data using create_forex_df function
data = create_forex_df()
print("The Table contains daily FOREX rates (against AUD) for the last 37 years.\n\n")
data.info()
```

```python
# Missing values by day of the week

fig, ax = plt.subplots()

# Plot the number of records on each day of week
plot = ax.bar(data.index.weekday_name.unique(), data.index.weekday.value_counts(sort=False))
ax.set_title('Number of Records each Day (Australian Western Standard Time)', fontsize=11)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(plot) 
plt.xticks(rotation=90)
plt.yticks([])
fig.tight_layout()
plt.ylim((0,2050))
plt.show()

```

#### Please be advised of the potential for illiquid market conditions particularly at the open of the trading week.

```python
def is_empty_record(weekday_id):
    
    record = data.loc[data.index.weekday==weekday_id].isnull().sum().sum()
    
    if record == len(data.columns):
        print("There are NO FOREX records for day {}.".format(weekday_id))
    else:
        print("There are {} FOREX records for day {}.".format(len(data.loc[data.index.weekday==weekday_id]), weekday_id))
        
for i in range(7):
    is_empty_record(i)
```

```python
# To ensure data quality for effective forecasting,
# we will drop the currency columns with more than 1000 missing values after 2010

# Reasons - Columns without most recent data aren't very effective for forecasting.
# INR, THB, etc. have many missing values, but have the most recent data, which can be possibly useful.

post_2010_data = data['2010':]
post_2010_nan = post_2010_data.isnull().sum()


currencies_to_drop = []

for currency, nan_count in post_2010_nan.items():
    
    if nan_count > 1000:
        currencies_to_drop.append(currency)

print(len(currencies_to_drop),"Currencies will be dropped =",currencies_to_drop)
```

```python
# Updated data
data_d = data.drop(currencies_to_drop, axis=1)

msno.matrix(data_d)
plt.hlines(6580, 0,20, color='yellow',linestyles='dashed', label='2010')
plt.show()
```

```python
# When is CNY missing
usd = data_d['USD']
mask = usd.isnull()

data_d.loc[usd[mask].index]

# We can see that when USD is missing, so is most of the other currencies in our 
# remaining dataset

# Probably Australian Holidays
```

```python
data_d[data_d['USD'].isnull()].index.max()

for currency in data_d.columns:
    null_mask = data_d[currency].isnull()
    
    last_missing_date = data_d[null_mask].index.max()
    
    print("Currency = {},\nYear of Last Missing Value = {}".format(currency, last_missing_date.year))
    print("Date = {}\n".format(last_missing_date))
```

#### Most data is available after 2010 (or least missing data after 2010).

```python
def plot_data_availability_by_year(data):   
   
    mask_nullity = data.isnull()
    
    nan_ids = data[mask_nullity].index    
    year = nan_ids.year

    data_with_nan = data.loc[nan_ids]

    data_available_per_year = data_with_nan.groupby(year).count().sum(axis=1)

    # Plot data availabilty per year
    plt.figure(figsize=(20,10))
    data_available_per_year.plot(kind='barh')
    plt.ylabel('Year', fontsize=16)
    plt.xlabel('Number of Non-Missing Values',fontsize=16)
    plt.title("Number of Non-Missing Values per Year",fontsize=16)
    plt.show()
    
plot_data_availability_by_year(data_d)
```
