```python
def get_dates_balance(text_file_name):
    
    with open(text_file_name, 'r') as f:
        content_all = f.read()
    
    date_pattern = r'[0-9][0-9] [A-Z][a-z][a-z]'
    pattern_balance = r'(\$[0-9]*,*[0-9]+.+[0-9]+ (C|D)R)'

    dates = []
    balance = []

    collector_flag = False
    
    for x in content_all.split("\n"):
        
        if "OPENING BALANCE" in x :
            collector_flag = True
        
        if "CLOSING BALANCE" in x :
            collector_flag = False
            
        if collector_flag:   
        
            if (x != "") & (len(re.findall(date_pattern, x)) > 0) & ~("CLOSING BALANCE" in x) & ~("OPENING BALANCE" in x):
                if 'Any' not in re.findall(date_pattern, x)[0]:
                    dates.append(re.findall(date_pattern, x)[0])

            if (x != "") & (len(re.findall(pattern_balance, x)) > 0) & ~("Closing Balance" in x):
                try :
                    balance.append(re.findall(pattern_balance, x)[0][0])
                except:
                    balance.append(re.findall(pattern_balance, x)[0])
                
    if len(dates) == len(balance) - 1:
        return dates, balance[1:]
    return dates, balance

d,b =get_dates_balance('Statement20190630.txt')
len(d), len(b)
```

```python
from glob import glob

filenames = glob("*.txt")
statement_periods = ["23 Jul 2018 - 31 Dec 2018", "1 Jan 2019 - 30 Jun 2019", "1 Jul 2019 - 31 Dec 2019", "1 Jan 2020 - 16 Mar 2020", "17 Mar 2020 - 30 Jun 2020"]
years = [2018, 2019, 2019, 2020, 2020]

for i, (fn, sp, yr) in enumerate(zip(filenames, statement_periods, years)):
    dt,bal = get_dates_balance(fn)
    df = pd.DataFrame({"Date":dt, "Balance":bal})
    
    df.loc[:,'Date'] = df['Date'].apply(lambda x : x + " " + str(yr))
    
    df.to_csv(("Bank Statement ({}).csv".format(sp)), index=False)
```
