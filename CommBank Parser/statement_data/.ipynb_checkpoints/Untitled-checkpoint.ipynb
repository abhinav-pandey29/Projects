```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import re
from glob import glob
```

```python
csv_filenames = glob("*.csv")

data = pd.DataFrame()
for csv_file in csv_filenames:
    df = pd.read_csv(csv_file, parse_dates=[0])
    data = pd.concat((data, df), axis=0)
    
data = data.sort_values('Date')
data.loc[:, 'Balance'] = data['Balance'].apply(lambda x : x.replace("$","").replace(",","").replace("CR","").replace("DR","")).astype(float)
data = data.set_index('Date')
data.info()
```

```python
data.plot()
```

```python
data.describe()
```

```python
std_df = data.copy()
std_df[std_df['Balance'] > 10000] = 5000

fig, ax = plt.subplots(figsize=(20,10))
std_df.plot(ax=ax)
plt.vlines('24-07-2019', -10, 10000, linestyle="--")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
data.plot(kind='hist',ax=ax, bins=600)
# plt.vlines('24-07-2019', -10, 1000, linestyle="--")
plt.xticks(range(0,35000, 1000), rotation=45)
plt.show()
```

```python
std_df = data.copy()
std_df[std_df['Balance'] > 10000] = 5000
std_df['year'] = std_df.index.year

fig, ax = plt.subplots(3,1,figsize=(10,8), sharex=True, sharey=True)
c = {2018:"red", 2019:"blue", 2020:"green"}
for i, yr in enumerate(std_df.year.unique()):
    sns.distplot(std_df[std_df['year']==yr]['Balance'], label=yr, bins=60, kde=True, color=c[yr], ax=ax[i])
    ax[i].legend()
plt.show()
```
